# AIエージェント：Claude Code と Computer Use

- Claude Code：ターミナルで動作するエージェント型コーディングアシスタント
- Computer Use：デスクトップ環境全体を操作できるツール群
- エージェントとして成功している理由の理解

## 1. このモジュールの概要

**取り上げる2つのアプリケーション：**

| アプリケーション | 動作環境 | できること |
|---|---|---|
| Claude Code | ターミナル | ファイル編集・バグ修正・開発ワークフロー支援 |
| Computer Use | デスクトップ全体 | Web閲覧・アプリ操作・視覚的UI操作 |

**これらがエージェントの好例である理由：**
- ツールの統合と使用
- 複数ステップのタスク実行
- 環境との相互作用
- 自律的な問題解決

## 2. Claude Code

ターミナルで動作するコーディングアシスタント。Claudeがターミナルに常駐して、あらゆるコーディング作業を支援する。

**提供されているツール群：**

| カテゴリ | 内容 |
|---|---|
| ファイル操作 | プロジェクト内のファイルの検索・読み取り・編集 |
| ターミナルアクセス | 会話から直接コマンドを実行 |
| Webアクセス | ドキュメント検索・コード例の取得 |
| MCPサーバーサポート | MCPサーバーに接続して追加ツールを拡張 |

**エージェントとしての視点：**
これらのツール群は、後で自分でエージェントを作る際のパターンそのもの。
「ファイルを読む → 内容を判断 → 編集する → 結果を確認」という複数ステップの自律実行がエージェントの基本。

**インストール（参考）：**
```bash
# Node.js が必要
npm install -g @anthropic-ai/claude-code

# 起動（初回はAnthropicアカウントへのログインが必要）
claude
```

## 4. 推奨ワークフロー

### 基本ワークフロー（3ステップ）
```
1. コンテキストを渡す   → 関連ファイルをClaudeに読ませる
2. 計画を立てる         → 「コードは書かず、アプローチだけ考えて」
3. 実装を依頼する       → 計画に基づいてコードを生成
```

### テスト駆動ワークフロー（より堅牢）
```
1. コンテキストを渡す          → 関連ファイルをClaudeに読ませる
2. テストケースを考えてもらう  → 新機能を検証するテストをブレスト
3. テストを書いてもらう        → 最も関連性の高いテストを選択して作成
4. テストに通るコードを書く    → 全テストをパスするまで反復
```

### 実例
```
> Read the math.py and document.py files          ← コンテキスト

> Plan to implement document_path_to_markdown:    ← 計画（コード不要）
  - ファイルパスを受け取る
  - ファイルの存在を検証する
  - 既存の binary_document_to_markdown を活用する

> Implement the plan                              ← 実装
```

## 6. Claude Code × MCP

Claude Code には MCPクライアントが組み込まれており、MCPサーバーを接続することで機能を拡張できる。

### MCPサーバーの登録
```bash
claude mcp add [server-name] [command-to-start-server]

# 例：ドキュメント処理サーバーを追加
claude mcp add documents uv run main.py
```
登録後は起動時に自動接続される。

### 人気のMCP連携

| MCPサーバー | 提供する機能 |
|---|---|
| `sentry-mcp` | Sentryのバグを自動検出・修正 |
| `playwright-mcp` | ブラウザ自動化・テスト |
| `figma-context-mcp` | FigmaデザインをClaudeに公開 |
| `mcp-atlassian` | Confluence・Jiraへのアクセス |
| `firecrawl-mcp-server` | Webスクレイピング |
| `slack-mcp` | Slackへのメッセージ投稿・返信 |

### 複数MCPサーバーを組み合わせたワークフロー例
```
Sentry  → 本番エラーの詳細を取得
  ↓
Jira    → 対応チケットの要件を読み取る
  ↓
Claude Code → コードを修正・テストを実行
  ↓
Slack   → 完了をチームに通知
```

MCPサーバーを組み合わせることで、既存の開発ツール・サービスとシームレスに連携するカスタムワークフローが作れる。

## 7. ワークフロー vs エージェント

### 使い分けの基準

| | ワークフロー | エージェント |
|---|---|---|
| **いつ使う** | 手順が事前に明確に定義できる場合 | どんなタスクが来るか事前にわからない場合 |
| **Claudeの役割** | 決められた手順を実行する | 目標とツールを与えられ、方法を自分で判断する |
| **例** | 画像→STEPファイル変換 | Claude Code（どんなコード作業が来るか不明） |

> このコースでツールを使ってClaudeにタスクの完了方法を判断させたのは、エージェントの例。

## 8. ワークフローパターン：評価器・最適化器

繰り返し改善が必要なタスクで使われるパターン。

### 構造
```
入力
  ↓
[プロデューサー] 出力を生成する（例：CadQueryでモデリング）
  ↓
[採点者] 出力を評価する（例：元画像と比較）
  ↓
合格？ → YES → 完了
        ↓ NO
      フィードバック → [プロデューサー] に戻る（反復）
```

### 画像→CAD変換ワークフローの例
```
① Claudeに画像を渡して物体を説明させる
② CadQueryライブラリでオブジェクトをモデリング
③ レンダリングを生成
④ レンダリングと元画像を比較・評価
   → 問題あり → ②に戻る
   → 合格    → STEPファイルを出力
```

### ポイント
- パターンを知ることで**再現可能な実装の指針**が得られる
- パターン自体が問題を解決するわけではなく、実際にコードで実装する必要がある
- 多くのエンジニアで実績があるパターンなので、自分のプロジェクトに応用できる

## 9. 並列化ワークフロー

### 単一プロンプトの問題点
複雑なタスクを1つのリクエストに詰め込むと：
- Claudeが複数の相反する要素を同時に考慮しなければならない
- 評価基準が混在して結果の信頼性が下がる
- 特定の部分だけを改善しにくい

### 並列化ワークフローの構造
```
入力（例：部品の画像）
  ↓　　↓　　↓　　↓　　↓　　↓
 金属  ポリマー セラミック 複合材 エラストマー 木材
（各素材に特化したプロンプトで並列評価）
  ↓　　↓　　↓　　↓　　↓　　↓
        集約ステップ
    （全評価を比較して最終推奨）
```

### メリット

| メリット | 内容 |
|---|---|
| 集中力 | Claudeが一度に1つの側面だけに集中できる |
| 最適化しやすい | 各サブタスクのプロンプトを独立して改善できる |
| 拡張しやすい | 新しい評価軸の追加 = 並列リクエストを1つ追加するだけ |
| 信頼性向上 | 認知負荷が減り、一貫性のある結果が得られる |

### いつ使うか
- 複数の独立した評価軸がある場合
- 複数の選択肢を比較させる場合
- 異なる専門分野が関わる意思決定をさせる場合

> **各並列サブタスクは独立して動作できる必要がある。** サブタスク同士が互いの結果に依存する場合は並列化できない。

## 10. ワークフローチェイニング

大きなタスクを**順序立てた小さなサブタスク**に分解して、順番に実行するパターン。

> 並列化との違い：並列化は独立したタスクを同時実行、チェイニングは依存関係のあるタスクを順番に実行。

### 例：ソーシャルメディア動画自動投稿ツール
```
① Twitterのトレンドを取得（外部API）
      ↓
② 最も興味深いトピックを選択（Claude）
      ↓
③ テーマについて調査（Claude）
      ↓
④ 短編動画の脚本を作成（Claude）
      ↓
⑤ AIアバターで動画を生成（外部ツール）
      ↓
⑥ ソーシャルメディアに投稿（外部API）
```
各ステップ間でLLM以外の処理（API呼び出しなど）を挟むこともできる。

### 長文プロンプト問題への対処

制約が多いと、Claudeは一部を無視しがちになる。

**悪い例（1つのプロンプトに詰め込む）：**
```
技術記事を書いて。AIが書いたと分かる表現はNG、絵文字はNG、
くだけた表現はNG、専門的なトーンで...（制約が多い）
→ Claudeが一部の制約を見落とす可能性がある
```

**良い例（チェイニング）：**
```
ステップ1：記事を生成する（制約違反があっても許容）
      ↓
ステップ2：修正に特化したプロンプトで改訂する
  「AIが書いたと分かる表現を削除、絵文字を削除、
   くだけた表現を技術的な文章に置き換えて」
```

### いつ使うか
- タスクに依存関係があり並列化できない場合
- 長いプロンプトでClaudeが制約を一部無視してしまう場合
- ステップ間で出力を検証・加工したい場合

## 11. ルーティングワークフロー

入力の種類を分類し、それぞれに特化した処理パイプラインに振り分けるパターン。

### 単一プロンプトの問題点
```
「プログラミング」→ 教育的なコンテンツが最適
「サーフィン」    → エンタメ性の高いコンテンツが最適
```
汎用プロンプトでは両方を高品質に扱えない。

### ルーティングの2ステップ
```
① 分類（ルーター）
   ユーザー入力 → Claudeに分類させる

② 特殊処理
   分類結果 → 対応するプロンプトテンプレートで生成
```

### アーキテクチャ
```
ユーザー入力
      ↓
  [ルーター] ← Claudeが分類
      ↓
 ┌────┬────┬────┬────┬────┬────┐
教育  娯楽 コメ Vlog レビ スト
育的  楽  ディ  グ   ュー リー
      ↓                        ↓
 各カテゴリに最適化されたパイプライン
```
入力は**全パイプラインではなく、1つの専用パイプラインにのみ**送られる。

### いつ使うか
- リクエストの種類によって処理方法を変えたい場合
- カテゴリを明確に定義できる場合
- カスタマーサービスBot・コンテンツ生成ツールなど

### 他のパターンとの比較

| パターン | 分岐 | 実行 |
|---|---|---|
| 並列化 | しない | 全サブタスクを同時実行 |
| チェイニング | しない | 順番に実行 |
| **ルーティング** | **する** | **該当パイプラインのみ実行** |

## 12. エージェント：ツールの組み合わせによる自律実行

### ワークフローとエージェントの再整理

| | ワークフロー | エージェント |
|---|---|---|
| 手順 | 事前に定義済み | Claudeが自分で決める |
| 柔軟性 | 低い | 高い |
| 信頼性 | 高い | 低め（トレードオフ） |
| コスト | 低い | 高め（トレードオフ） |

### ツールの組み合わせで複雑なタスクを解決する

シンプルなツールでも組み合わせることで複雑な要求に対応できる。

```
【日時ツール3つ】
get_current_datetime / add_duration_to_datetime / set_reminder

「今何時？」             → get_current_datetime のみ
「11日後は何曜日？」     → get_current_datetime → add_duration_to_datetime
「来週水曜にジムのリマインダー」 → 3つを順番に組み合わせ
「90日保証はいつまで？」 → 購入日を確認してから計算（追加情報が必要と自律判断）
```

### ツールは抽象的であるべき

**悪い例（過度に専門化）：**
```
refactor_code() / install_dependencies() / run_tests()
→ 想定外のシナリオに対応できない
```

**良い例（適度に抽象的）：**
```
bash / read / write / edit / glob / grep
→ Claudeが基本ツールを組み合わせて、あらゆるシナリオに対応
```

Claude Code はこの原則の実例。「コードのリファクタリング」専用ツールはなく、汎用ツールを組み合わせて対応する。

### 設計のポイント
- **組み合わせ可能なツール**を提供する
- 同じツールセットで**シンプルなワークフローにも複雑なインタラクションにも**対応できる
- ユーザーのフィードバックに応じてアプローチを変えられる → 固定ワークフローでは難しい

## 13. 環境の観察（Environment Inspection）

Claudeは盲目的に行動するため、**アクションの結果を観察できる仕組み**を与えることが重要。

### なぜ必要か
```
Claudeがボタンをクリックした
  → 新しいページに移動した？メニューが開いた？エラーになった？
  → 結果を観察できないと次のアクションを判断できない
```

### 基本原則：書く前に読む
```
ファイルを変更する前に → 必ず現在の内容を読む
UI操作の後に         → スクリーンショットで結果を確認する
API呼び出しの後に     → レスポンスに期待するデータが含まれているか確認する
```

### システムプロンプトで観察を促す例
```
動画生成エージェントのシステムプロンプト：
- whisper.cpp でセリフの位置をタイムスタンプ付きで確認せよ
- FFmpeg でスクリーンショットを一定間隔で抽出して視覚的に検査せよ
- 生成コンテンツを元の要件と比較せよ
```

### 観察できることで改善される点

| 改善点 | 内容 |
|---|---|
| 進捗把握 | タスク完了まであとどれくらいか把握できる |
| エラー処理 | 予期しない結果を検知して修正できる |
| 品質保証 | 完了とみなす前に出力を検証できる |
| 適応行動 | 観察結果に基づいてアプローチを変えられる |

### 設計時の自問
> **「このアクションが成功したかどうかをClaudeはどのように判断するのか？」**

観察手段を用意していないエージェントは、盲目的に命令を実行するだけになる。

## 14. ワークフロー vs エージェント：まとめと使い分け

### メリット・デメリット比較

| | ワークフロー | エージェント |
|---|---|---|
| **精度** | 高い（1つのサブタスクに集中） | 低め |
| **信頼性** | 高い・予測可能 | 低め・予測しにくい |
| **テスト・評価** | しやすい（手順が明確） | 難しい（手順が不定） |
| **柔軟性** | 低い（特定タスク専用） | 高い（想定外にも対応） |
| **UX** | 制約がある | 自由度が高い |
| **事前設計** | 多く必要 | 少なくて済む |

### 使い分けの指針

```
明確なプロセスがある → ワークフロー
予測不可能・多様なリクエスト → エージェント
```

### 重要な原則
> **まずワークフローで実装し、エージェントは本当に必要な場合のみ使う。**

ユーザーが求めているのは「高度なエージェント」ではなく「常に安定して動作する製品」。信頼性・予測可能性を優先するのが運用アプリケーションの基本。

## 15. パターンまとめ

| パターン | 概要 | 使いどころ |
|---|---|---|
| **評価器・最適化器** | 生成 → 評価 → NG なら修正のループ | 品質基準を満たすまで反復改善したい |
| **並列化** | 独立したサブタスクを同時実行 | 複数の視点で独立して評価・比較したい |
| **チェイニング** | 前のステップの結果を次に渡す順次実行 | 手順に依存関係がある・長いプロンプトを分割したい |
| **ルーティング** | 入力を分類して専用パイプラインに振り分け | 入力の種類によって処理方法を変えたい |
| **エージェント** | ツールと目標を与えてClaudeが自律判断 | 手順が事前に決められない・多様なリクエストに対応したい |

**基本方針：まずワークフローパターンで実装できないか検討し、難しい場合のみエージェントを使う。**
